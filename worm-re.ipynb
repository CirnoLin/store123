```python
# coding=gbk
from bs4 import BeautifulSoup as bs
import openpyxl
import urllib.request
import re
import json
import demjson
import time
```

```python
def url(u):
    
    import ssl
    context = ssl._create_unverified_context()
    
    url ='https://yjszs.ecnu.edu.cn/system/sszszyml_list.asp'
    headers= {'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3',
              'Accept-Language': 'zh-CN,zh;q=0.9',
              'Cache-Control': 'max-age=0', 'Connection': 'keep-alive',
              'Cookie': 'ASPSESSIONIDAARSRDDC=KEIFPHMDJKBGPANCJCFMNBGG',
              'Host': 'yjszs.ecnu.edu.cn',
              'Sec-Fetch-Mode': 'navigate',
              'Sec-Fetch-Site': 'none',
              'Sec-Fetch-User': '?1',
              'Upgrade-Insecure-Requests': '1', 'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36'}

    req = urllib.request.Request(u,headers = headers)
    html = urllib.request.urlopen(req,context = context)
    html = html.read()

    return html
```

```python
def tr():
    u = 'https://yjszs.ecnu.edu.cn/system/sszszyml_list.asp'
    ur =[]
    html = url(u)
    html = bs(html,'lxml')
    cont = html.findAll('tr',valign="middle")
    for i in cont:
        ii = i.findAll('td',valign="center")[1]
        u = 'https://yjszs.ecnu.edu.cn/system/'+ii.a.attrs['href']
        ur.append([u,i.a.text])
        print(i.a.text)
        print('\n')
    return ur
```

```python
def excel(shu):
    wb = openpyxl.Workbook()
    word = wb.active
    word['A1'] = '序号'      
    word['B1'] = '招生年份'        
    word['C1'] = '招生专业'        
    word['D1'] = '学位类型'        
    word['E1'] = '招生人数'        
    word['F1'] = '研究方向'        
    word['G1'] = '考试科目'        
    word['H1'] = '备注'
    for x in range(1,len(shu)+1):
        A = 'A{0}'.format(x+1)
        B = 'B{0}'.format(x+1)
        C = 'C{0}'.format(x+1)
        D = 'D{0}'.format(x+1)
        E = 'E{0}'.format(x+1)
        F = 'F{0}'.format(x+1)
        G = 'G{0}'.format(x+1)
        H = 'H{0}'.format(x+1)
        word[A] = shu[x-1][0]       
        word[B] = shu[x-1][1]        
        word[C] = shu[x-1][2]        
        word[D] = shu[x-1][3]        
        word[E] = shu[x-1][4]        
        word[F] = shu[x-1][5]       
        word[G] = shu[x-1][6]       
        word[H] = shu[x-1][7]
    name = '华东师范大学2020年硕士研究生招生专业目录'
    wb.save('{0}.xlsx'.format(name))
    print('保存完成'+name)
    print('\n')
```

```python
def main():
    urllist = tr()#获取学系链接
    ex = []
    for i in urllist:
        name = i[1]#学系名字
        html = url(i[0])#表格内容
        html = bs(html,'lxml')
        cont = html.findAll('tr',valign="middle")
        #ex = [] #列表保存内容
        for x in cont:
            content=[]#列表保存每一行内容
            list2 = x.findAll('td')
            for h in list2:
                hh = h.text
                hh = re.sub('\r','',hh)
                hh = re.sub('\t','',hh)
                hh = re.sub('\n','',hh)
                #print(hh)
                content.append(hh)
            ex.append(content)        
    excel(ex)#生成表格
```

```python
if __name__ == '__main__':
    main()
```

```python
json_all  =[]
def read_07_Excel(file_path):
    wb = openpyxl.load_workbook(file_path)  #打开文件
    sheet = wb["Sheet"]  #通过sheet名称锁定表格
    # new_list =[]
    for row in sheet.rows:  #循环所有的行
        list_release = []
        for cell in row:#循环行中所有的单元格
            print(cell.value, "\t", end="")  #获取单元格的值
            list_release.append(cell.value)
        print()
        if list_release[0]!="序号":
            process_to_json(list_release)
```

```python
def process_to_json(json_list):
    list_json = json_list
    release_dirt = {'序号':'test','招生年份':'test','招生专业':'test','学位类型':'test','招生人数':'test','研究方向':'test','考试科目':'test','备注':'test'}
    release_dirt['序号'] = list_json[0]
    release_dirt['招生年份'] = list_json[1]
    release_dirt['招生专业'] = list_json[2]
    release_dirt['学位类型'] = list_json[3]
    release_dirt['招生人数'] = list_json[4]
    release_dirt['研究方向'] = list_json[5]
    release_dirt['考试科目'] = list_json[6]
    release_dirt['备注'] = list_json[7]
    print(release_dirt)
    # json_str = json.dumps(release_dirt)
    json_all.append(release_dirt)
```

```python
def process_json():
    json_dirt = {"": json_all}
    json_str = json.dumps(json_dirt)
    f = open('华东师范大学2020年硕士研究生招生专业目录.json', 'w')
    f.write(json_str)
    f.close()
```
